```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import make_scorer
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,shuffle = True)
```

# Unchained Multioutput Regressor

### Predicting the geometry -- Gradient Boosting Decision Tree Regressor

```python
reg = MultiOutputRegressor(GradientBoostingRegressor(learning_rate = 0.1,n_estimators = 300, max_depth = 5, max_features = 501))
reg.fit(X_train.values,y_train.values)
```

```python
r2_score(y_test,reg.predict(X_test), multioutput = 'variance_weighted')
```

### Predicting the Transmission values for the wavelength range =1000 - 2000 nm - GBDTR

```python
reg2 = MultiOutputRegressor(GradientBoostingRegressor(learning_rate = 0.1,n_estimators = 100, max_depth = 5))
reg2.fit(y_train,X_train)
```

```python
r2_score(X_test,reg2.predict(y_test), multioutput = 'variance_weighted')
```

```python
reg.predict(reg2.predict([[40,1700]]))
```

### Predicting the geometry -- Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
reg3 = MultiOutputRegressor(RandomForestRegressor(n_estimators = 100, max_depth = 5,max_features = 501))
reg3.fit(X_train,y_train)
```

```python
r2_score(y_test,reg3.predict(X_test), multioutput = 'variance_weighted')
```

### Predicting the transmission values -- Random Forest Regressor

```python
reg4 = MultiOutputRegressor(RandomForestRegressor(n_estimators = 100, max_depth = 3))
reg4.fit(y_train,X_train)
```

```python
r2_score(X_test,reg4.predict(y_test), multioutput = 'variance_weighted')
```

# Chained Multioutput Regressor
